```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras import regularizers
```

```python
video_base = "/content/drive/MyDrive/deepfake_project/dataset"

fake_videos_path = os.path.join(video_base, "fake")
real_videos_path = os.path.join(video_base, "real")

output_path = "/content/drive/MyDrive/deepfake_project/frames"

fake_folder = os.path.join(output_path, "fake")
real_folder = os.path.join(output_path, "real")

os.makedirs(fake_folder, exist_ok=True)
os.makedirs(real_folder, exist_ok=True)
```

```python
def get_video_files(folder):
    videos = []
    for root, dirs, files in os.walk(folder):
        for file in files:
            if file.endswith(('.mp4', '.avi', '.mov', '.mkv')):
                videos.append(os.path.join(root, file))
    return videos

fake_videos = get_video_files(fake_videos_path)
real_videos = get_video_files(real_videos_path)

print("Fake videos:", len(fake_videos))
print("Real videos:", len(real_videos))
```

```python
def is_blurry(image, threshold=60):   # كان 100
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return cv2.Laplacian(gray, cv2.CV_64F).var() < threshold

def is_too_bright(image, threshold=250):  # كان 230
    return image.mean() > threshold
```

```python
def extract_clean_faces(video_path, output_folder, label, max_faces=100):
    face_cascade = cv2.CascadeClassifier(
        cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
    )

    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    saved = 0

    while cap.isOpened() and saved < max_faces:
        ret, frame = cap.read()
        if not ret:
            break

        # 🔥 أقل تخطي → فريمات أكثر
        if frame_count % 3 != 0:
            frame_count += 1
            continue

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.2, 4)  # تحسين detection

        if len(faces) == 0:
            frame_count += 1
            continue

        for (x, y, w, h) in faces[:1]:
            face = frame[y:y+h, x:x+w]

            if face is None or face.size == 0:
                continue

            face = cv2.resize(face, (224, 224))

            # فلترة أخف
            if is_blurry(face):
                continue
            if is_too_bright(face):
                continue

            filename = f"{label}_{os.path.basename(video_path)}_{saved}.jpg"
            filepath = os.path.join(output_folder, filename)

            cv2.imwrite(filepath, face)
            saved += 1

        frame_count += 1

    cap.release()
```

```python
import os

print("Checking if faces already extracted...")

# إذا الفولدرات موجودة وفيها صور → لا تعيد الاستخراج
if os.path.exists(fake_folder) and os.path.exists(real_folder) and \
   len(os.listdir(fake_folder)) > 0 and len(os.listdir(real_folder)) > 0:

    print("Faces already extracted ✔")

else:
    print("Extracting CLEAN faces...")

    for video in fake_videos:
        extract_clean_faces(video, fake_folder, 'fake', max_faces=100)

    for video in real_videos:
        extract_clean_faces(video, real_folder, 'real', max_faces=100)

    print("Done ✔")
```

```python
fake_count = len(os.listdir(fake_folder))
real_count = len(os.listdir(real_folder))
total      = fake_count + real_count

print(f"\nFake frames : {fake_count}")
print(f"Real frames : {real_count}")
print(f"Total       : {total}")

# FIX 1: Class Weights — يعوّض الخلل بين الـ Classes
weight_fake = total / (2 * fake_count)
weight_real = total / (2 * real_count)
class_weights = {0: weight_fake, 1: weight_real}

print(f"\nClass weights → fake: {weight_fake:.2f} | real: {weight_real:.2f}")
```

```python
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=10,
    zoom_range=0.1,
    horizontal_flip=True
)

train_data = datagen.flow_from_directory(
    output_path,
    target_size=(224,224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_data = datagen.flow_from_directory(
    output_path,
    target_size=(224,224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
total = fake_count + real_count
weight_fake = total / (2 * fake_count)
weight_real = total / (2 * real_count)

class_weights = {0: weight_fake, 1: weight_real}

print(class_weights)
```

```python
base_model = MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3)
)

for layer in base_model.layers:
    layer.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.4)(x)
x = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=x)

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

model.fit(
    train_data,
    validation_data=val_data,
    epochs=15,
    class_weight=class_weights,
    callbacks=[early_stop]
)
```

```python
base_model.trainable = True

for layer in base_model.layers[:-50]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# FIX 3: EarlyStopping في Fine-tuning كمان
early_stop_ft = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

print("\n--- Phase 2: Fine-tuning ---")
model.fit(
    train_data,
    validation_data=val_data,
    epochs=10,
    callbacks=[early_stop_ft],    # FIX 3
    class_weight=class_weights    # FIX 1
)
```

```python
loss, acc = model.evaluate(val_data)
print(f"Model Accuracy:{acc*100:.2f}%")
```
